```python
#add the Technical Indicators to the data
Stocks=[NET, DKNG, LNG, BRO, UBER, LRCX, INVA, FBP]
import talib as ta
for ticker in Stocks:
  ticker['SMA_10_close']=ta.SMA(ticker['Close'],10)
  ticker['SMA_21_close']=ta.SMA(ticker['Close'],20)
  ticker['SMA_50_close']=ta.SMA(ticker['Close'],50)
  ticker['SMA_200_close']=ta.SMA(ticker['Close'],200)
  ticker['EMA_10_close']=ta.EMA(ticker['Close'],10)
  ticker['EMA_20_close']=ta.EMA(ticker['Close'],20)
  ticker['EMA_50_close']=ta.EMA(ticker['Close'],50)
  ticker['EMA_200_close']=ta.EMA(ticker['Close'],200)
  ticker['DEMA_10_close']=ta.DEMA(ticker['Close'],10)
  ticker['DEMA_20_close']=ta.DEMA(ticker['Close'],20)
  ticker['DEMA_50_close']=ta.DEMA(ticker['Close'],50)
  ticker['DEMA_200_close']=ta.DEMA(ticker['Close'],200)
  ticker['TEMA_10_close']=ta.TEMA(ticker['Close'],10)
  ticker['TEMA_20_close']=ta.TEMA(ticker['Close'],20)
  ticker['TEMA_50_close']=ta.TEMA(ticker['Close'],50)
  ticker['TEMA_200_close']=ta.TEMA(ticker['Close'],200)
  ticker['RSI Close']=ta.RSI(ticker['Close'],14)
  ticker['AvgTrueRange']=ta.ATR(ticker['High'],ticker['Low'],ticker['Close'])
```

```python
# added Kaufman and momentum
for ticker in Stocks:
  ticker['KAMA_10_close']=ta.KAMA(ticker['Close'],10)
  ticker['KAMA_20_close']=ta.KAMA(ticker['Close'],20)
  ticker['KAMA_50_close']=ta.KAMA(ticker['Close'],50)
  ticker['Momentume_Close']=ta.MOM(ticker['Close'])
```

```python
#add macd using the ema and bollinger bands

def macd(ticker):
  ema_26=ta.EMA(ticker['Close'],26)
  ema_12=ta.EMA(ticker['Close'],12)
  ticker['macd']=ema_12-ema_26

def bollinger_bands(ticker):
  window=21
  num_stds=2
  rollingmean=ticker.Close.rolling(window).mean()
  rollingstd=ticker.Close.rolling(window).std()

  ticker['bbband_high']=(rollingmean + (rollingstd*num_stds)).fillna(0)
  ticker['bbband_low']=(rollingmean - (rollingstd*num_stds)).fillna(0)

  for ticker in Stocks:
    macd(ticker)
    bollinger_bands(ticker)

```

```python

def bestmodel(ticker):
  import numpy as np
  import tensorflow as tf
  from keras.models import Sequential
  from keras.layers import LSTM,Bidirectional
  from keras.layers import Dense
  from keras.layers import Dropout
  import pandas as pd
  from matplotlib import pyplot as plt
  from sklearn.preprocessing import StandardScaler
  import seaborn as sns
  import pandas as pd
  import yfinance as yf
  from tensorflow.keras.callbacks import EarlyStopping
  from sklearn.metrics import classification_report, confusion_matrix
  import keras_tuner as kt

  # work on the NU dataset and then make a funciton
  import datetime as dt
  from datetime import datetime
  #convert dates and check dates
  ticker.reset_index(inplace=True)
  ticker['Date'] =ticker['Date'].dt.strftime('%Y/%m/%d')
  all_dates=pd.to_datetime(ticker['Date'])

  # move close column to be first column
  def swap_columns(df, col1, col2):
      col_list = list(df.columns)
      x, y = col_list.index(col1), col_list.index(col2)
      col_list[y], col_list[x] = col_list[x], col_list[y]
      df = df[col_list]
      return df

  ticker=swap_columns(ticker,'Close','Open')

  #variables for training
  cols=list(ticker)[1:]

  data=ticker[cols].astype(float)

  #standerdize data
  from sklearn.preprocessing import StandardScaler
  scaler=StandardScaler()
  df_scaled_training=scaler.fit_transform(data)

  trainX=[]
  trainY=[]
  import numpy as np
  n_future=1
  n_past=14

  for i in range(n_past, len(df_scaled_training) - n_future +1):
      trainX.append(df_scaled_training[i - n_past:i, 0:df_scaled_training.shape[1]])
      trainY.append(df_scaled_training[i + n_future - 1:i + n_future, 0])

  trainX,trainY=np.array(trainX),np.array(trainY)

  # hyperparamter tuning - layers(refer to the githubcode)
  def build_model(hp):
    model = Sequential()
    model.add(LSTM(hp.Int('input_unit',min_value=50,max_value=1000,step=50),return_sequences=True, input_shape=(trainX.shape[1], trainX.shape[2])))
    for i in range(hp.Int('n_layers', 1, 4)):
        model.add(LSTM(hp.Int(f'lstm_{i}_units',min_value=50,max_value=1000,step=50),return_sequences=True))
    model.add(LSTM(6))
    model.add(Dropout(hp.Float('Dropout_rate',min_value=0,max_value=0.8,step=0.1)))
    model.add(Dense(6))
    model.add(Dropout(hp.Float('Dropout_rate',min_value=0,max_value=0.8,step=0.1)))
    model.add(Dense(1))

    learning_rate_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=1e-2,
        decay_steps=10000,
        decay_rate=0.7)
    optimizers = tf.keras.optimizers.Adam(learning_rate=learning_rate_schedule)

    model.compile(loss='mean_squared_error', optimizer=optimizers,metrics = ['mse'])
    return model

    loss = model.history['loss']
    val_loss = model.history['val_loss']
    epochs_range = range(100)
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()

  tuner= kt.RandomSearch(
      build_model,
      objective='mse',
      max_trials=3,
      executions_per_trial=5,
      overwrite=True
      )
  tuner.search(
      x=trainX,
      y=trainY,
      epochs=100,
      validation_split=0.3,
      batch_size=64,
  )


  best_model = tuner.get_best_models(num_models=1)[0]
  best_model.summary()
  tuner_model=best_model.fit(trainX, trainY,epochs=100,verbose=1,validation_split= 0.2)

  n_futures=90
  forecast_period_dates = pd.date_range(list(all_dates)[-15], periods=n_futures, freq='1d').tolist()
  print(forecast_period_dates)
  forecast = best_model.predict(trainX[-n_futures:])
  prediction_copies = np.repeat(forecast, data.shape[1], axis=-1)
  y_pred_future = scaler.inverse_transform(prediction_copies)[:,0]
  forecast_dates = []
  for time_i in forecast_period_dates:
    forecast_dates.append(time_i.date())
  testdates_prices=[]
  for time_i in all_dates:
      testdates_prices.append(time_i.date())
  df_test = pd.DataFrame({'Date':np.array(forecast_dates), 'Close':y_pred_future})
  df_test['Date']=pd.to_datetime(df_test['Date'])
  return df_test
```
